```python
import numpy as np 
import matplotlib.pyplot as plt 
a = 0.2
b = 0.2 
c = 5.7 
x0, y0, z0 = 0.0, 1.0, 0.0 
T = 250.0 
dt = 0.01 
num_steps = int(T/dt) 
time_value = np.arange(0, T + dt, dt) 
x_value = np.zeros(num_steps + 1) 
y_value = np.zeros(num_steps + 1) 
z_value = np.zeros(num_steps + 1) 
x_value[0], y_value[0], z_value[0] = x0, y0, z0 
for item in range(num_steps): 
    dxdt = -y_value[item] - z_value[item] 
    dydt = x_value[item] + a * y_value[item] 
    dzdt = b + z_value[item] * (x_value[item] - c) 

    x_value[item + 1] = x_value[item] + dxdt * dt 
    y_value[item + 1] = y_value[item] + dydt * dt 
    z_value[item + 1] = z_value[item] + dzdt * dt 

plt.plot(x_value, y_value) 
```
